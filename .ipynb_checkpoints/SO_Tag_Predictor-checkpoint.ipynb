<h2>2.2 Mapping the real-world problem to a Machine Learning Problem </h2>

<h3> 2.2.1 Type of Machine Learning Problem </h3>

<p> It is a multi-label classification problem  <br>
<b>Multi-label Classification</b>: Multilabel classification assigns to each sample a set of target labels. This can be thought as predicting properties of a data-point that are not mutually exclusive, such as topics that are relevant for a document. A question on Stackoverflow might be about any of C, Pointers, FileIO and/or memory-management at the same time or none of these. <br>
__Credit__: http://scikit-learn.org/stable/modules/multiclass.html
</p>

<h3>2.2.2 Performance metric </h3>

<b>Micro-Averaged F1-Score (Mean F Score) </b>: 
The F1 score can be interpreted as a weighted average of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0. The relative contribution of precision and recall to the F1 score are equal. The formula for the F1 score is:

<i>F1 = 2 * (precision * recall) / (precision + recall)</i><br>

In the multi-class and multi-label case, this is the weighted average of the F1 score of each class. <br>

<b>'Micro f1 score': </b><br>
Calculate metrics globally by counting the total true positives, false negatives and false positives. This is a better metric when we have class imbalance.
<br>

<b>'Macro f1 score': </b><br>
Calculate metrics for each label, and find their unweighted mean. This does not take label imbalance into account.
<br>

https://www.kaggle.com/wiki/MeanFScore <br>
http://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html <br>
<br>
<b> Hamming loss </b>: The Hamming loss is the fraction of labels that are incorrectly predicted. <br>
https://www.kaggle.com/wiki/HammingLoss <br>

```python
# Countvectorizer uses white spaces, punctuation marks, etc for tokenizing the text
# However in our case we want the words to be separated by only space exclusively
# so we use 'split()' will tokenize each tag using space.
vectorizer = CountVectorizer(tokenizer = lambda x: x.split())

# fit_transform() does two functions: First, it fits the model
# and learns the vocabulary; second, it transforms our training data
# into feature vectors. The input to fit_transform should be a list of strings.
tag_dtm = vectorizer.fit_transform(tag_data['Tags'])
```

```python
#'get_feature_name()' gives us the vocabulary.
tags = vectorizer.get_feature_names()

#Lets look at the tags we have.
print("Some of the tags we have :", tags[:10])
```

```python
# http://www.sqlitetutorial.net/sqlite-python/create-tables/

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by db_file
    :param db_file: database file
    :return: Connection object or None
    """
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
 
    return None

def create_table(conn, create_table_sql):
    """ create a table from the create_table_sql statement
    :param conn: Connection object
    :param create_table_sql: a CREATE TABLE statement
    :return:
    """
    try:
        c = conn.cursor()
        c.execute(create_table_sql)
    except Error as e:
        print(e)

def create_database_table(database, query):
    conn = create_connection(database)
    if conn is not None:
        create_table(conn, query)
    else:
        print("Error! cannot create the database connection.")
    conn.close()
```

#### Create a table QuestionsProcessed with the following fileds:
* question - the body of the question with code removed
* tags - the tags in the question (output variable in the prediction problem)
* words_pre - no. of words in 'question' before processing
* words_post - no. of words in 'question' after processing
* is_code - a binary flag indicating whether the quesion contained code or not

```python
sql_create_table = """CREATE TABLE IF NOT EXISTS QuestionsProcessed (question text NOT NULL, tags text, words_pre integer, words_post integer, is_code integer);"""
create_database_table("Processed.db", sql_create_table)
```

<h3> 4.1 Preprocessing of questions </h3>

<ol> 
    <li> Separate Code from Body </li>
    <li> Remove Spcial characters from Question title and description (not in code)</li>
    <li> <b> Give more weightage to title : Add title three times to the question </b> </li>
   
    <li> Remove stop words (Except 'C') </li>
    <li> Remove HTML Tags </li>
    <li> Convert all the characters into small letters </li>
    <li> Use SnowballStemmer to stem the words </li>
</ol>

```python
def striphtml(data):
    """Function to remove HTML tags from data"""
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, ' ', str(data))
    return cleantext

# Set of stopwords
stop_words = set(stopwords.words('english'))

# Initialize the Stemmer
stemmer = SnowballStemmer("english")
```

### Pre processing of 0.5M datapoints and inserting in a new database "Processed.db"

```python
# http://www.bernzilla.com/2008/05/13/selecting-a-random-row-from-an-sqlite-table/
start = datetime.now()
preprocessed_data_list=[]

#reader.fetchone()
questions_with_code=0 # Counter maintaining the no. of questions which contain code
len_pre=0  # Length of the question before processing
len_post=0  # Length of the question after processing
questions_processed = 0

# Extract the rows from the reader one by one
for row in reader:
    # Binary Flag denoting whether the question conatains code
    is_code = 0
    
    # Extract individual fields from the database row
    title, question, tags = row[0], row[1], str(row[2])
    
    # Check if "<code>" is present in the question
    if '<code>' in question:
        questions_with_code+=1
        is_code = 1
    x = len(question)+len(title)
    len_pre+=x
    
    # Find the parts of the question containing code
    # This can also be inserted into the table for analysis but we avoid it
    code = str(re.findall(r'<code>(.*?)</code>', question, flags=re.DOTALL))
    
    # Remove code from the question
    question=re.sub('<code>(.*?)</code>', '', question, flags=re.MULTILINE|re.DOTALL)
    # Remove HTML tags from the question
    question=striphtml(question.encode('utf-8'))
    # Encode the "title" in utf-8
    title=title.encode('utf-8')
    
    # adding title three time to the data to increase its weight
    # add tags string to the training data
    question=str(title)+" "+str(title)+" "+str(title)+" "+question

    # Remove all characters from question except: A to Z, a to z, 0 to 9, '#', '+', '.', '/', '-'
    question=re.sub(r'[^A-Za-z0-9#+./-]+',' ',question)
    
    # Convert everything to lower case and tokenize the words using white space
    # 'words' is the list of all words in the question
    words = question.lower().split()
    
    # Removing all stopwords and single letters from question except for the letter 'c' and apply stemming
    question=' '.join(str(stemmer.stem(j)) for j in words if j not in stop_words and (len(j)!=1 or j=='c'))
    
    len_post+=len(question)
    
    # Finally create a tuple of all fields to be inserted in the Processed.db
    tup = (question,tags,x,len(question),is_code)
    questions_processed += 1
    writer.execute("insert into QuestionsProcessed(question,tags,words_pre,words_post,is_code) values (?,?,?,?,?)",tup)
    
    # Print after every 50000 insertion to track progress
    if (questions_processed%50000==0):
        print("number of questions completed=",questions_processed)

# Calculate avg length of question before and after processing - Just for Statistics
avg_len_pre=(len_pre)/questions_processed
avg_len_post=(len_post)/questions_processed

print( "Avg. length of questions(Title+Body) before processing: %d"%avg_len_pre)
print( "Avg. length of questions(Title+Body) after processing: %d"%avg_len_post)
print ("Percent of questions containing code: %d"%((questions_with_code*100)/questions_processed))

print("Time taken to run this cell :", datetime.now() - start)
```

__ We will sample the number of tags instead of considering all of them (due to limitation of computing power) __

__ Selecting 500 Tags __

```python
def tags_to_choose(n):
    # Count no. of times each tag appears and put the counts in a list
    t = multilabel_y.sum(axis=0).tolist()[0]  # Column sum of the countvectorizer object
    
    # Get the indices of the tags in the sorted order of their counts (descending order)
    sorted_tags_i = list(np.argsort(-np.array(t)))
    # Alternate way
    '''sorted_tags_i = sorted(range(len(t)), key=lambda i: t[i], reverse=True)'''
    
    # Get the rows top n Tags from the ouput of CountVectorizer
    multilabel_yn=multilabel_y[:,sorted_tags_i[:n]]
    
    return multilabel_yn

def questions_explained_fn(n):
    # Get the top n tags
    multilabel_yn = tags_to_choose(n)
    
    # Count the no. of tags present in each row(question)
    x = multilabel_yn.sum(axis=1)
    
    # Return only those rows with at least ont tag present (from the top n tags)
    return (np.count_nonzero(x==0))
```

#### Check percentage questions are covered by considering different no. of tags
This is similar to the variance explained in PCA

```python
questions_explained = []
total_tags=multilabel_y.shape[1]
total_qs=preprocessed_data.shape[0]
for i in range(500, total_tags, 100):
    questions_explained.append(np.round(((total_qs-questions_explained_fn(i))/total_qs)*100,3))
```

```python
fig, ax = plt.subplots()
ax.plot(questions_explained)
xlabel = list(500+np.array(range(-50,450,50))*50)
ax.set_xticklabels(xlabel)
plt.xlabel("Number of tags")
plt.ylabel("Number Questions coverd partially")
plt.grid()
plt.show()
# you can choose any number of tags based on your computing power, minimun is 500(it covers 90% of the tags)
print("with ",5500,"tags we are covering ",questions_explained[50],"% of questions")
print("with ",500,"tags we are covering ",questions_explained[0],"% of questions")
```
